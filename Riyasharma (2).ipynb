```python
def col_convert(df):
    id_vars = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code']
    value_vars = df.columns.difference(id_vars).tolist()
    df = pd.melt(df, id_vars=id_vars, value_vars=value_vars, var_name='Year', value_name='Value')
    df['Year'] = pd.to_datetime(df['Year'], format='%Y')
    return df['Year']

def datafram_convert(df):
  Year = col_convert(df)

  return df['Country Name'] , Year
```

```python
path = "/content/API_AG.LND.FRST.ZS_DS2_en_csv_v2_5358376.csv"
df = pd.read_csv(path)
Country_Name, Year = datafram_convert(df)

print(Country_Name)
print(Year)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv("/content/API_AG.LND.FRST.ZS_DS2_en_csv_v2_5358376.csv")
df.columns
```
